```python
import gensim
from urllib.request import urlopen
import numpy as np
from bs4 import BeautifulSoup
import nltk
import ssl
```

```python
wiki = "http://en.wikipedia.org/wiki/"
titles = ["Integral", "Riemann_integral", "Riemann-Stieltjes_integral", "Derivative",
    "Limit_of_a_sequence", "Edvard_Munch", "Vincent_van_Gogh", "Jan_Matejko",
    "Lev_Tolstoj", "Franz_Kafka", "J._R._R._Tolkien"]
```

```python
context = ssl._create_unverified_context()
```

```python
# return bs4.element.ResultSet
def parse_old(url):
    x = urlopen(url, context=context)
    x = x.read()
    x = BeautifulSoup(x)
    x = x.find("div",id="bodyContent")
    x = x.find_all("p")
    return x
```

```python
# return list of str
def parse(url):
    x = urlopen(url, context=context)
    x = x.read()
    x = BeautifulSoup(x,)
    x = x.find("div",id="bodyContent")
    x = x.find_all("p")
    paragraphs = []
    for p in x:
        paragraphs.append(str(p))
    return paragraphs
```

```python
articles_o = [parse(url) for url in [wiki+x for x in titles]]
```
