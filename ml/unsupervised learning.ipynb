#### Dane 1

```python
points = np.concatenate([rnd.multivariate_normal((0,0), np.array([[1,0],[0,1]]), 20), 
                         rnd.multivariate_normal((5,2), np.array([[1,0],[0,1]]), 20),
                         rnd.multivariate_normal((2,7), np.array([[1,0],[0,1]]), 20),])
```

#### Wyznaczanie ilosci klastrow: Inercja

```python
points = np.concatenate([rnd.multivariate_normal((0,0), np.array([[1,0],[0,1]]), 20), 
                         rnd.multivariate_normal((5,2), np.array([[1,0],[0,1]]), 20),
                         rnd.multivariate_normal((2,7), np.array([[1,0],[0,1]]), 20),])
plt.scatter(points[:,0], points[:,1])
plt.show()
```

```python
import seaborn as sns
colors = np.array(sns.color_palette("Set2", 20))
```

```python
ks = range(1, 8)
inertia = []

for k in ks:
    model = KMeans(k)
    model.fit(points)
    inertia.append(model.inertia_)
    plt.subplot(2,4,k)
    plt.scatter(points[:,0], points[:,1])
    plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1],c="yellow",s=70)
plt.show()

plt.plot(range(1,8), inertia)
plt.show()
```

```python
# reguÅ‚a Å‚okcia. sami oceniamy gdie konczy sie najwiekszy spadek i jest wypÅ‚aszczenie
```

#### Wyznaczanie ilosci klastrow: Miara Silhouette 

```python
from sklearn.metrics import silhouette_score
```

```python
ks = range(2, 9)
sil = []

for k in ks:
    model = KMeans(k)
    model.fit(points)
    sil.append(silhouette_score(points, model.labels_))
    plt.subplot(2,4,k)
    plt.scatter(points[:,0], points[:,1])
    plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1],c="yellow",s=70)
plt.show()

plt.plot(range(2,9), sil)
plt.show()
```
