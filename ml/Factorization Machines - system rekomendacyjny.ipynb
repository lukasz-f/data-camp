```python
from pandas import pivot_table
import numpy as np
ratings_pivot = pivot_table(ratings, values='rating', index=['user_id'], columns=['movie_id'], aggfunc=np.count_nonzero)
ratings_pivot
```

```python
ratings_pivot_sum = ratings_pivot.apply(lambda row: row / row.sum(), axis=1)
ratings_pivot_sum
```

```python
ratings_pivot_sum_reset = ratings_pivot_sum.reset_index()
ratings_pivot_sum_reset
```

```python
ratings_ = pd.merge(ratings, ratings_pivot_sum_reset, how='left', on=['user_id'])
ratings_
```
