```python
# Analiza ilościowa (text mining) - do grupowania i klasyfikacji tekstów
# Analiza jakościowa - procesowanie języka naturalnego (NLP)
```

```python
wiki = "http://en.wikipedia.org/wiki/"
titles = ["Integral", "Riemann_integral", "Riemann-Stieltjes_integral", "Derivative",
    "Limit_of_a_sequence", "Edvard_Munch", "Vincent_van_Gogh", "Jan_Matejko",
    "Lev_Tolstoj", "Franz_Kafka", "J._R._R._Tolkien"]
```

```python
from urllib.request import urlopen
import numpy as np
from bs4 import BeautifulSoup
import nltk
```

```python
def pars(url):
    x = urlopen(url)
    x = x.read()
    x = BeautifulSoup(x)
    x = x.find('div', id='bodyContent')
    x = x.find_all('p')
    return x
```

```python
articles = [pars(x) for x in [wiki+title for title in titles]]
```

```python
articles
```

```python
articles[0]
```

```python
import re
articles = [re.sub('<.+?>', " ", str(a)) for a in articles]
articles
```

```python
articles = [re.sub('\{.+?\}', " ", str(a)) for a in articles]
articles
```

```python
articles = [nltk.word_tokenize(a) for a in articles]
articles[0]
```

```python
import string
articles = [[w for w in a if w not in string.punctuation] for a in articles]
articles[0]
```

```python
stemer = nltk.PorterStemmer()
# stemming - zamienia słowa na quasi-rdzenie
stemer = articles = [ [w.lower() for w in a] for a in articles]
# articles = [[stemer.stem(w) for w in a] for a in articles]
articles[0]
```

```python
articles = [[w for w in a if w not in nltk.corpus.stopwords.words('english')] for a in articles]
# stopwords - and, or, because, should, etc.
articles[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
# CountVectorizer przeprowadza zaawansowaną wektoryzację słów (wszystkie powyższe operacje w jednej funkcji)
# n-gramy - zbitki n słów traktować jako tokeny ngram_range - min/max (1 pojedyncze słowa, 2 zbitki dwóch słów, etc)
# max_df/min_df - słowo występuje maksymalnie/minimalnie w danej liczbie dokumentów
```

```python
articles = [" ".join(a) for a in articles]
articles
```

```python
?CountVectorizer
# zamiana na wektory liczności występowania
```

```python
c = CountVectorizer()
dtm = c.fit(articles)
art = dtm.transform(articles)
art
```

```python
dtm.vocabulary_ # słownik indeksów kolumn z wyrazami
```

```python
dtm.get_feature_names()
```

```python
art.getcol(dtm.get_feature_names().index('kingdom')).todense()
```

```python
# funkcja zwracająca najczęściej występujące słowa w artykułach

def top_words(M,dtm,k):
    words = np.array(dtm.get_feature_names())
    print(words)
    return(np.array([words[np.squeeze(np.array(np.argsort(M[i,:].todense())))[-k:]] for i in range(M.shape[0])]))

top_words(art,dtm,6)
```

# TF-IDF - określanie ważności słów
<hr>

```python
# Metoda TF-IDF waży słowa na podstawie dwóch czynników: term frequency i inverse document frequency
# TF - Słowo jest tym ważniejsze, im częściej pojawia się w dokumencie
# IDF - odwrotność częstości występowania w całej strukturze danych - jeśli pojawia się wszędzie często - nie jest ważne
```

```python
# tej metody można użyć nie tylko do analizy tekstu - dzięki niej można zaobserować ważność występowania
# danej cechy u konkretnego użytkownika vs cała zbiorowość - np. mogłoby się okazać, że najczęściej
# czytaną na danym portalu kategorią jest polityka - i występuje ona jako najczęstsza dla wszystkich
# użytkowników. Samo badanie per użytkownik nie odda zatem zróżnicowania - trzeba zanalizować na tle całej
# zbiorowości (odwrotność występowania w całym zbiorze)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# zamiana na wektory ważności - ważone wg TFIDF
```

```python
tfidf = TfidfVectorizer()
art = tfidf.fit_transform(articles)
art.todense()[:,:5] # wagi słów (w pierwszych 5 kolumnach)
```

```python
top_words(art,tfidf,5)
# słowa o największych wagach (poprzednio najliczniej występujące)
```
