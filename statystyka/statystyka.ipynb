```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
plt.plot(np.array([0, 1, 2, 3]), [4, 3, 2, 1])
plt.show()
```

```python
plt.scatter(np.arange(4), np.arange(4, 0, -1))
plt.show()
```

```python
data = np.array([-2.1, -1.3, -0.4, 1.9, 5.1, 6.2])
sns.kdeplot(data, kernel='gau', bw='silverman')
plt.show()
```

```python
plt.plot(stats.cumfreq(data,numbins=25)[0])
plt.show()
```

```python
p = 0.4
bernoulliDist = stats.bernoulli(p)
trials = bernoulliDist.rvs(100)
plt.hist(trials)
plt.show()
```

```python
mean, variance, skew, kurtosis = stats.bernoulli.stats(p, moments='mvsk')
print(mean)
print(variance)
print(skew)
print(kurtosis)
```

```python
x = np.arange(0, 2)
pmf = stats.bernoulli.pmf(x, p)
#pmf = bernoulliDist.pmf(x)
plt.plot(x, pmf, 'bo', ms=8, label='bernoulli pmf')
plt.vlines(x, 0, pmf, colors='b', lw=5, alpha=0.5)
plt.show()
```

```python
x = np.arange(-1, 2,0.01)
plt.plot(x, bernoulliDist.cdf(x), 'bo--', ms=8, label='bernoulli cdf')
plt.show()
```

```python
lambda1 = 2
poissonDist = stats.poisson(lambda1)
x = np.arange(0, 10)
plt.plot(x, poissonDist.pmf(x), 'bo', ms=8, label='poisson pmf')
plt.vlines(x, 0, poissonDist.pmf(x), colors='b', lw=5, alpha=0.5)
plt.show()
```

```python
t = np.arange(-5, 5, 0.05)
uniformDist=stats.uniform(-1,2)
plt.plot(t, uniformDist.pdf(t), '--', ms=8, label='frozen pdf')
plt.show()
```

```python
plt.plot(t, uniformDist.cdf(t), '--', ms=8, label='bernoulli cdf')
plt.show()
```
