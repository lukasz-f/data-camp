# Box-Muller

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
```

# Zadanie
Proszę wylosować dane z rozkładu Normalnego metodą Boxa-Mullera.

Porsze na jednym wykresie narysować próbkę oraz gęstość tego rozkładu. 

```python
n = 1000
u1 = np.random.random(n)
u2 = np.random.random(n)
r_squared = -2*np.log(u1)
r = np.sqrt(r_squared)
theta = 2*np.pi*u2
x = r*np.cos(theta)
y = r*np.sin(theta)
```

```python
x_plot = np.linspace(-4, 4)
plt.hist(x, bins=100, normed=True, linewidth=2)
plt.plot(x_plot, stats.norm.pdf(x_plot), linewidth=2)
plt.axis([-4, 4, 0, 0.6])
plt.title('Histogram of exponential PRNGs')
plt.show()
```
