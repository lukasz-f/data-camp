# Rozkład chi kwadrat z $n$ stopniami swobody
Rozkład chi kwadrat ma gęstość:

$$
f_{X}(x) = 
\left\{ \begin{array}{ll}
{\frac {1}{2^{\frac {n}{2}}\Gamma \left({\frac {n}{2}}\right)}}\;x^{{\frac {n}{2}}-1}e^{-{\frac {x}{2}}} 
   & \textrm{gdy } x \geq 0\\
0 & \textrm{gdy } x < 0 
\end{array} \right. ,
$$

gdzie $\Gamma$ oznacza funkcję Gamma https://pl.wikipedia.org/wiki/Funkcja_\%CE\%93


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

# Definiujemy zmienną losową

```python
t = np.arange(0, 20, 0.05)
n=4
X=stats.chi2(n)
```

# Narysujmy gętość zmiennej losowej

```python
fig, ax = plt.subplots(1, 1)
plt.plot(t, X.pdf(t), '--', ms=8, label='frozen pdf')
plt.show()
```

# Narysujmy dystrybuantę zmiennej losowej

```python
fig, ax = plt.subplots(1, 1)
ax.plot(t, X.cdf(t), '--', ms=8, label='bernoulli cdf')
plt.show()
```

# Wylosujmy próbkę i narysujmy histogram na jednym obrazku

```python
r = X.rvs(size=1000)
fig, ax = plt.subplots(1, 1)
ax.hist(r, normed=True, histtype='stepfilled', alpha=0.5, label='histogram')
ax.legend(loc='best', frameon=False)
ax.plot(t, X.pdf(t), 'k-', lw=2, label='a=-1, b=1')
ax.legend()
plt.show()
```

# Narysuj gęstości rozkładu jednostajnego z różnymi parametrami

```python
n_values = [1, 2, 3]
fig, ax = plt.subplots(1, 1)
for n in n_values:
    dist = stats.chi2(n)
    plt.plot(t, dist.pdf(t), ls="--",label=r'$n=%1.f' % n)  
plt.legend()
plt.show()
```

# Narysuj kilka histogramów różnych próbek z roskładu jednostajnego o tych samych parametrach.

```python
fig, ax = plt.subplots(nrows=5, ncols=5, figsize=(10, 10))

for i in np.arange(0,5):
    for j in np.arange(0,5):
        r = X.rvs(size=100)
        ax[i,j].hist(r, normed=True)
plt.show()
```
