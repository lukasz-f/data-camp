# Rozkład normalny
Gęstość zmiennej losowej $X$ o rozkładzie normalnym z parametrami $\mu$ i $\sigma$ jest dana przez 
$$
f(x) = \frac{1}{\sigma \sqrt{2 \pi} } e^{-\frac{(x-m)^2}{2\sigma^2} }.
$$

Definiujemy zmienną losową

```python
(mu, sigma) = (0, 1)
normDist = stats.norm(mu, sigma)
t = np.arange(-5, 5, 0.05)
```

Narysujmy gętość zmiennej losowej

```python
# Display the probability mass function (``pmf``):
fig, ax = plt.subplots(1, 1)
plt.plot(t, normDist.pdf(t), '--', ms=8, label='frozen pdf')
ax.set_xlim([-5,5])
ax.set_ylim([-0.1,1])
plt.show()
```

Narysujmy dystrybuantę zmiennej losowej

```python
fig, ax = plt.subplots(1, 1)
ax.plot(t, normDist.cdf(t), '--', ms=8, label='bernoulli cdf')
plt.show()
```

Narysujmy gęstości rozkładu jednostajnego z różnymi parametrami

```python
sigma_values = [0.5, 1.0, 2.0]
mu_values = [0, 1, 2]
fig, ax = plt.subplots(1, 1)
for mu, sigma in zip(mu_values, sigma_values):
    dist = stats.norm(mu, sigma)
    plt.plot(t, dist.pdf(t), ls="--", c='black', label=r'$a=%1.f,\ b=%.1f$' % (mu, mu+sigma))
ax.set_xlim([-5,5])
ax.set_ylim([-0.1,1.1])    
plt.legend()
plt.show()
```

Narysujmy kilka histogramów różnych próbek z roskładu normalnego o tych samych parametrach.

```python
fig, ax = plt.subplots(nrows=5, ncols=5, figsize=(10, 10))

for i in np.arange(0,5):
    for j in np.arange(0,5):
        r = normDist.rvs(size=100)
        ax[i,j].hist(r, normed=True)
plt.show()
```
