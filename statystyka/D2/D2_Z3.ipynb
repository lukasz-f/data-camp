# Rozkład Poissona
Rozkład prawdopodobieństwa zmiennej losowej $X$ o rozkładzie Poissona z parametrem $\lambda > 0$ wyraża się wzorem:
$$
P(X = k) = e^{-\lambda} \cdot \frac{\lambda^{k}}{k!}.
$$


```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
```

Zdefiniuj zmienną losową o rozkładzie Poissona  $\lambda=2$

```python
lambda1 = 2
poissonDist = stats.poisson(lambda1)
```

Narysuj gęstość zmiennej losowej o rozkładzie Poissona

```python
fig, ax = plt.subplots(1, 1)
# Display the probability mass function (``pmf``):
ax.set_xlim([-1,10])
ax.set_ylim([0,1])
x = np.arange(0, 10)
ax.plot(x, stats.poisson.pmf(x, lambda1), 'bo', ms=8, label='poisson pmf')
ax.vlines(x, 0, stats.poisson.pmf(x, lambda1), colors='b', lw=5, alpha=0.5)
plt.show()
```

Wylosuj próbkę i narysuj histogram

```python
trials = poissonDist.rvs(100)
trials
plt.hist(trials)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
# Display the probability mass function (``pmf``):
ax.set_xlim([-1,10])
ax.set_ylim([-0.2,1.2])

x = np.arange(-10, 10,0.01)
ax.plot(x, stats.poisson.cdf(x, lambda1), 'bo--', ms=8, label='poisson cdf')
plt.show()
```

Policz momenty

```python
mean, var, skew, kurt = poissonDist.stats(moments='mvsk')
print(mean)
print(var)
print(skew)
print(kurt)
```

Narysuj gętość zmiennej losowej dla różnych parametrów

```python
# Arbitrarily select 3 lambda values
lambdas = [1,4,10]

k = np.arange(20)       # generate x-values
markersize = 8
for par in lambdas:
    plt.plot(k, stats.poisson.pmf(k, par), 'o--', label='$\lambda={0}$'.format(par))

# Format the plot
plt.legend()
plt.title('Poisson distribution')
plt.xlabel('X')
plt.ylabel('P(X)')
plt.show()
```
