```python
from sklearn import datasets

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Iris Data Set

```python
iris = datasets.load_iris()
print('type(iris):', type(iris))
print('iris.keys():', iris.keys())
print('type(iris.data):', type(iris.data))
print('type(iris.target):', type(iris.target))
print('iris.data.shape:', iris.data.shape)
print('iris.feature_names:', iris.feature_names)
print('iris.target_names:', iris.target_names)
```

```python
X = iris.data
y = iris.target
iris_df = pd.DataFrame(X, columns=iris.feature_names)
iris_df.head()
```

```python
pd.scatter_matrix(iris_df, c=y, figsize=(10, 10), s=100, marker='D')
plt.show()
```

#### Congressional Voting Records Data Set 

```python
# ../datasets/voting-records/house-votes-84.data
votes = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/voting-records/house-votes-84.data', \
                    names=['party', 'infants', 'water', 'budget', 'physician', 'salvador', \
                           'religious', 'satellite', 'aid', 'missile', 'immigration', 'synfuels', \
                           'education', 'superfund', 'crime', 'duty_free_exports', 'eaa_rsa' \
                          ])
votes.head(10)
```

```python
votes.info()
```

```python
plt.figure()
sns.countplot(x='education', hue='party', data=votes, palette='RdBu')
plt.xticks([0,1,2], ['Yes', 'No', 'Not Vote'])
plt.show()
```

```python
plt.figure()
sns.countplot(x='satellite', hue='party', data=votes, palette='RdBu')
plt.xticks([0,1,2], ['No', 'Yes', 'Not Vote'])
plt.show()
```

```python
plt.figure()
sns.countplot(x='missile', hue='party', data=votes, palette='RdBu')
plt.xticks([0,1,2], ['No', 'Yes', 'Not Vote'])
plt.show()
```
