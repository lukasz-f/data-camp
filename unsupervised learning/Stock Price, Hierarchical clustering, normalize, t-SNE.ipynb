```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.manifold import TSNE
```

## Company Stock Price Movements Dataset

```python
stocks = pd.read_csv('https://assets.datacamp.com/production/course_2072/datasets/company-stock-movements-2010-2015-incl.csv', 
                 index_col=0)
stocks.head()
```

```python
movements = stocks.values
companies = stocks.index.values
movements.shape, companies.shape
```

## Hierarchical clustering of Stocks

```python
# SciPy hierarchical clustering doesn't fit into a sklearn pipeline.
# Use the normalize() function instead of Normalizer.
# Normalize the movements
normalized_movements = normalize(movements)
```

```python
# Calculate the linkage
mergings = linkage(normalized_movements, method='complete')
```

```python
# Plot the dendrogram
plt.figure(figsize=(20,10))
dendrogram(mergings,
           labels=companies,
           leaf_rotation=90,
           leaf_font_size=12)
plt.show()
```

## t-SNE map

```python
# Create a TSNE instance
model = TSNE(learning_rate=50)
```

```python
# Apply fit_transform to normalized_movements
tsne_features = model.fit_transform(normalized_movements)
```

```python
# Select features
xs = tsne_features[:,0]
ys = tsne_features[:,1]
```

```python
# Scatter plot
plt.figure(figsize=(20,10))
plt.scatter(xs, ys, alpha=0.5)

# Annotate the points
for x, y, company in zip(xs, ys, companies):
    plt.annotate(company, (x, y), fontsize=14, alpha=0.75)

plt.show()
```
