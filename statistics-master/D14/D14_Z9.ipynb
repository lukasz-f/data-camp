# Wielowymiarowy rozkład normalny

```python
import matplotlib.pylab as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.mlab as mlab
from matplotlib import cm
```

# Narysujmy gęstość dwuwymiarowego rozkładu Normalnego

```python
X = np.arange(-3, 3, 0.01)
Y = np.arange(-3, 3, 0.01)
X, Y = np.meshgrid(X, Y)
cov = np.array([[ 0.6744121 , -0.16938146], [-0.16938146,  0.21243464]])
Z = mlab.bivariate_normal(X, Y,  np.sqrt(cov[0, 0]), np.sqrt(cov[1, 1]), 0, 0, cov[1, 0])
fig = plt.figure(figsize=(14,6))
ax = fig.add_subplot(1, 2, 1, projection='3d')
_ = ax.plot_surface(X, Y, Z, linewidth=0.2, cmap=cm.viridis)
plt.show()
```

# Narysujmy kontury gęstości dwuwymiarowego rozkładu Normalnego

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))
ax.set_ylim(-1.5, 1.5)
ax.set_xlim(-2, 2)
levels = np.arange(0.0, 0.5, 0.01)
_ = plt.contour(X, Y, Z, levels, origincmap=cm.viridis)
plt.show()
```

# Wylosujmy próbkę z dwuwymiarowego rozkładu Normalnego

```python
mean = [0, 0]
cov = np.array([[ 0.6744121 , -0.16938146], [-0.16938146,  0.21243464]])
samples = np.random.multivariate_normal(mean, cov, 1000)
w, v = np.linalg.eig(cov)
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))
_ = plt.scatter(samples[:,0], samples[:,1])
plt.show()
```
