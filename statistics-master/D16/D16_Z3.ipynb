```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

Fs = 150.0;  # sampling rate
Ts = 1.0/Fs; # sampling interval
t = np.arange(0,1,Ts) # time vector

ff1 = 5;   # frequency of the signal
ff2 = 14;   # frequency of the signal
y = np.sin(2*np.pi*ff1*t)+np.sin(2*np.pi*ff2*t)


fig, ax = plt.subplots(3, 1,figsize=(10,6))
ax[0].plot(t,np.sin(2*np.pi*ff1*t))
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')
ax[1].plot(t,np.sin(2*np.pi*ff2*t),'r') # plotting the spectrum
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')
ax[2].plot(t, y,'g') # plotting the spectrum
ax[2].set_xlabel('Time')
ax[2].set_ylabel('Amplitude')
plt.show()
```

```python
n = len(y) # length of the signal
k = np.arange(n)
T = n/Fs
frq = k/T # two sides frequency range
frq = frq[range(np.int(n/2))] # one side frequency range

Y = np.fft.fft(y)/n # fft computing and normalization
z = np.fft.ifft(Y)*n # ifft computing and normalization

Y = Y[range(np.int(n/2))]
```

```python
fig, ax = plt.subplots(3, 1,figsize=(10,6))
ax[0].plot(t,y)
ax[0].set_xlabel('Time')
ax[0].set_ylabel('Amplitude')
ax[1].plot(frq,abs(Y),'r') # plotting the spectrum
ax[1].set_xlabel('Freq (Hz)')
ax[1].set_ylabel('|Y(freq)|')
ax[2].plot(t, z,'g') # plotting the spectrum
ax[2].set_xlabel('Time')
ax[2].set_ylabel('Amplitude')
plt.show()
```
