```python
x = np.arange(100)
y1 = np.random.normal(0,1,100).cumsum()
y2 = np.random.normal(0,1,100).cumsum()
plt.plot(x, np.c_[y1,y2])
plt.show()
```

```python
np.random.seed(123)
x = np.arange(100)
y1 = np.random.normal(0,1,100).cumsum()
y2 = np.random.normal(0,1,100).cumsum()
plt.plot(x, y1, "k-", x,y2,"k--")
plt.show()
```

```python
np.random.seed(123)
x = np.arange(100)
y1 = np.random.normal(0,1,100).cumsum()
y2 = np.random.normal(0,1,100).cumsum()
plt.plot(x, y1, "k-", label="wykres1")
plt.plot(x,y2,"k--", label="wykres2")
plt.axis("tight")
plt.legend()
plt.show()
```

```python
np.random.seed(123)
x = np.arange(100)
y1 = np.random.normal(0,1,100).cumsum()
y2 = np.random.normal(0,1,100).cumsum()
a1, = plt.plot(x, y1, "k-")
a2, = plt.plot(x, y2,"k--")
plt.axis("tight")
plt.legend([a1,a2], ["wykres1", "wykres2"])
plt.show()
```
