```python
# Narysuj koło o środku w punkcie (0, 0) i promieniu 1. 
# Powinno być ono  wypełnione półprzezroczystą barwą czerwoną, 
# mieć silnie zarysowany brzeg i nie wyglądać jak elipsa.

t = np.linspace(0, 2*np.pi, 100, endpoint=False)
plt.axis("equal")
plt.ylim(-1.1, 1.1)
plt.fill(np.cos(t), np.sin(t), facecolor=(1, 0, 0, 0.1),
    edgecolor="darkred", linewidth=5)
plt.text(0, 0, "(0,0)", verticalalignment="top")
plt.plot(0, 0, "ko")
plt.show()
```

```python
t = np.linspace(0, 2*np.pi, 100, endpoint=False)
x = 16*np.sin(t)**3
y = 13*np.cos(t) - 5*np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)
plt.fill(x+20, y, color='k', alpha=0.1)
plt.fill(x+10, y+15, color='k', alpha=0.5)
plt.fill(x, y, color='k', alpha=0.8)
plt.text(0, 0, s="0.8", color="w", horizontalalignment='center', verticalalignment='center')
plt.text(10, 15, s="0.5", color="w", horizontalalignment='center', verticalalignment='center')
plt.text(20, 0, s="0.1", color="k", horizontalalignment='center', verticalalignment='center')
plt.axis("off")
plt.show()
```

```python
# Narysuj wykres funkcji sin(x2)na przedziale [0, 2].
# Zamieść na osi Ox etykiety postaci 0,..., 2,..., 2. 
# Umieść legendę o białym tle w lewym dolnym rogu.
x = np.linspace(0, 2*np.pi, 1000)
y = np.sin(x**2)
plt.plot(x, y, color="darkorange", label=r"$\sin(x^2)$")

l = plt.legend(loc="lower left", shadow=True, frameon=True)
l.get_frame().set_facecolor("white")

plt.xticks(np.linspace(0, 2*np.pi, 5), [r"$0$", r"$\frac{\pi}{2}$",
    r"$\pi$", r"$\frac{3}{2}\pi$", r"$2\pi$"])

plt.show()
```
