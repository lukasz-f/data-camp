```python
def notowania(spolka):
    f = requests.get("https://stooq.pl/q/?s=%s" % spolka)
    return re.findall(r'<span id=aq_%s_c2>([\d\.]+)</span>' % spolka, f.text)
```

```python
notowania('cdr')
```

```python
notowania('mbk')
```

```python
f = requests.get("https://stooq.pl/q/?s=cdr")
match = re.search(r'Ostatnia sesja(.*?)</td>', f.text)
hrefs = re.findall(r'href=[\'"]?([^\'" >]+)', match.group())
hrefs = ['https://stooq.pl/' + href for href in hrefs]
hrefs
```

```python
for href in hrefs:
    f = requests.get(href)
    wyniki = re.findall(r">Kurs.*?(\d+\.\d+)<", f.text)
    print(wyniki)
```
