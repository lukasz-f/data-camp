```python
result.head()
```

```python
mean_rating = result[["Rating"]].groupby(result["Title"]).apply(np.mean)
```

### F: wykonaj poprzedni punkt, odrzucając wcześniej filmy które nie uzyskały wystarczająco dużo głosów (np 100)

```python
mean_rating_cnt = list(result.Title)
```

```python
cnt = Counter(mean_rating_cnt)
cnt
```

```python
movies_addmited = [x for x in cnt if cnt[x] >= 100]
movies_addmited
```

```python
#mean_rating.index
mean_rating_pro = [ (y,x[0]) for x, y in zip(mean_rating.values, mean_rating.index)  if y  in movies_addmited]
```

```python
mean_rating_pro = sorted(mean_rating_pro,key=lambda tup: tup[1], reverse=True)
print("Most interesting film: ",mean_rating_pro[0][0]," rate: ",mean_rating_pro[0][1])
```
